# Feature Engineering

For this dataset I don't think much has to be done, but lets combine the marketing campaign participation and number of kids/teens into a single variable of the combined total. We also need to make dummy variables for Marital Status and Education.

```python
df['TotalCmp'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] #total number of marketing campaigns the customer took adavantage of
df['TotalDependents'] = df['Kidhome'] + df['Teenhome'] #total number of kids in the home

df.drop(columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Kidhome', 'Teenhome'], inplace=True)
df.shape
```

We have now modified our dataset to only have 21 features instead of the original 29 by combining some features and deleting some others. However, we our feature count is going to increase again because the KMeans method we are going to be using later requires all data to be numerical. This means we have to find a way to transform our two non-numerical features, Marital Status and Education, into a numerical format. One way we can do that is with a method called one hot encoding, in which each unique value in the non-numerical feature gets its own column and is filled with either a 1 if the customer chose that value or 0 if they chose a different value.

The pandas .get_dummies() method will do this for us.

```python
df_final = pd.get_dummies(df, columns=['Marital_Status', 'Education'])
df_final.info()
```

One last thing we need to do is scale our data. If, we do not scale our data, the features will be unequally weighted in the KMeans model we are about to use to segment our customers, such that it will split the data into groups based on the variable with the highest variance, in our case income. We care about the variance because we will use the sum of the square distance between the center of a customer segment cluster (physical point in high dimensional space) and a customers representative numerical location. The greater the variance of the varaible/ feature the greater the impact on our sum of squares.

```python
mm = MinMaxScaler().fit(df_final)
df_mm = pd.DataFrame(mm.transform(df_final), columns=df_final.columns)
```

# Customer Segmentation

Now that we have our data prepared, we can finally start answering some of our questions. We will use KMeans clustering method to determing for our customer segments. The KMeans method works as follows...

First, choose the number of data clusters you think the data might have and the KMeans method will randomly choose a starting position for the center of each of those clusters
Second, the distance from the cluster centers to the every customers numerically represented point is calcualted
Third, closest points to each cluster center is determined
Fourth, the cluster center is moved to the center (average value) of every customer point that is closest to that cluster center
Fifth, repeart the second through fourth steps

Since we were not told how many customer segments we should have, we will have to determine that number of clusters for ourselves. We can do this by looking at the trend between the number of clusters we choose and the sum of the squared error, often called inertia. What we are looking for in the rend is a sharp turn, often called the "elbow", in which the inertia stops decreasing rapidly as the number of clusters increases and starts to decrease slowly. We can do this by fitting multiple KMeans models with varying number of clusters and looking at the tradeoff.

```python
sum_square = []

for n in range(1,30):
    km = KMeans(n_clusters=n, random_state=11)
    km.fit(df_mm)
    sum_square.append(km.inertia_)

fig = plt.figure(figsize=(12,8))
plt.plot(range(1,30), sum_square, marker='o')
plt.ylabel('Sum of Squares')
plt.xlabel('Number of Clusters')
plt.title('Sum of Squared Distances vs. Number of Clusters')
plt.grid(True)

plt.show()
```

It is difficult to tell where the elbow is because we typially want a sharp bend. However, it looks like the elbow occurs at around 8 or 9 clusters due to the tapering off of the sum of squares. We can also try to qualitatively determine the number of clusters using the t-distributed stocastic neighbor embedding (t-SNE) method. t-SNE is a dimensionality reducing algorithm that maintainins the relative distance between points in high-dimensional space.

```python
tsne_feats = TSNE(learning_rate=100, n_components=2).fit_transform(df_mm)

fig = plt.figure(figsize=(15,10))
plt.scatter(tsne_feats[:,0], tsne_feats[:,1])
plt.xlabel('TSNE Feature 1')
plt.ylabel('TSNE Feature 2')

plt.show()
```

In the above graph we can ignore the X and Y labels because they don't represent anything meaningful after the transformation. Also, there looks to be 8 to 9 main clusters and then there are a lot of small clusters spread out near the center of the graph. Let's choose 8 clusters for now since people tend to be more alike than dissimiliar. We will now make another KMeans model using 8 clusters since we believe it to be accurate - Thus allowing us to answer our first question.

```python
#Create the 8 clusters
km = KMeans(n_clusters=8, random_state=11).fit(df_mm)
df_clustered = df_mm.copy()
df_clustered['Grouping'] = km.labels_
```

Now that we grouped all of the customer into 1 of 8 customer segments, let's start looking at some of their details. Also, just to make sure their were no abnormalities in the algorithm let's make sure none of the clusters has only a couple of associated data points.

```python
#Distribution of customers in each cluster
df_clustered['Grouping'].value_counts()
```

The customer segments have a relatively even distribution of customers, although group 0 has approximately double the amount in some other groupings. Now let's look at each groups characteristics using bar graphs.

```python
df_clustered.groupby('Grouping').mean().plot(kind='bar', figsize=(25,14))
plt.ylabel('Percentage of Customers that have the Attribute')
```

Even though we can not tell by the color due repeats, we can see that each group has it's own unique fingerprint based on the attributes with a value greater than 0.55 and also values equal to zero because the attributes that they score high on are just as important as the attributes that they score low on. Let's look closer at these distinguishing attributes by calculating the average of each feature for each customer segment and showing the features each segment scored high on - Which allows us to answer our second question.

```python
df_clustered['Grouping'] = km.labels_
avg_atts = df_clustered.groupby('Grouping').mean() #Average of attributes

avg_atts[avg_atts > 0.55].dropna(axis=1, how='all')
```

So our groupings are based on education level and relationship status. We have a group of 

Group 0: married people that have graduated highschool  
Group 1: single people that have graduated highschool  
Group 2: people that identify as being together that graduated highschool  
Group 3: married people with PhDs  
Group 4: people that identify as being together and have higher education  
Group 5: single people with higher education  
Group 6: married people that obtained a master's degree  
Group 7: those that are divorced  

Let's see if we can answer our third question by looking at these segments unique shopping habits.

```python
df_final['Grouping'] = km.labels_
avg_atts = df_final.groupby('Grouping').mean() #Average of attributes

avg_atts[avg_atts > 0.55].dropna(axis=1, how='all')

#What they buy
axis_font = {'size':'14'}
title_font = {'size':'18'}

with plt.style.context('dark_background'):
    avg_atts[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].plot(kind='bar', 
                                                                                                                       stacked=True, 
                                                                                                                       figsize=(15,10))
    plt.xlabel('Group Number', **axis_font)
    plt.ylabel('Average Money Spent', **axis_font)
    plt.title('Group Average Spend per Category', **title_font)
    plt.xticks(rotation=0)
    plt.savefig('GroupAvgSpend.png')
    
```

Overall, there does not seem to be much variation in some of the spending habits. People that are married and have PhDs tend to spend more on wine, but not much on everything else. Looks like in general, for people whose highest level of education is highschool tend to spend more on gold and fish products, no matter if they are single or in a relationship (groups 1, 2, and 3).

We can confirm our hypothesis by doing a bootstrap statistical analyis in which we resample the customers without a PhD to create a new dataset of which we can calulate their average spend on wine. If we do this multiple times we will get a distribution of potential mean spending on wine of customers without a PhD, which can then be compared to how much married people with PhDs spend on wine on average. We can also calculate an approximate significane value

```python
#Test PhD theory - null hypothesis is PhDs spend the same amount of money on wine as everyone without a PhD
phds = df[df['Education'] == 'PhD']
other = df[df['Education'] != 'PhD']

avg_wine_spend = []
for i in range(10000):
    avg_wine = (np.random.choice(other['MntWines'], replace=True, size=len(other))).mean() #avg wine spend of bootstrap
    avg_wine_spend.append(avg_wine)
```

We just create 10,000 datasets resampling the customers that are not married with a PhD. Let's calculate how many of these sample datasets had a mean wine spend equal to or greater than those that are married with PhDs.

```python
(pd.DataFrame(avg_wine_spend) > phds['MntWines'].mean()).sum()
```

There were 0 datasets out of 10,000 in which the other customers had a mean wine spend equal to married people with PhDs, so our theory was correct. This is very significant statistically and to the business since we can target people with similar characteristics in marketing campaigns selling wine. Let's amke a quick visualization too, to see just how different their average is.

```python
phd_avg_wine = phds['MntWines'].mean()

pd.DataFrame(avg_wine_spend, columns=['Bootstrap ']).hist(bins=100, figsize=(15,10))
plt.plot([phd_avg_wine, phd_avg_wine], [0, 350])
```

One last thing we should do is look at the customer segments shopping habits to see if there are any significant differences.

```python
axis_font = {'size':'14'}
title_font = {'size':'18'}

with plt.style.context('dark_background'):
    avg_atts[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'TotalCmp']].plot(kind='bar', 
                                                                                                                       stacked=True, 
                                                                                                                       figsize=(15,10))
    plt.xlabel('Group Number', **axis_font)
    plt.ylabel('Shopping Habits', **axis_font)
    plt.title('Group Shopping Habits per Category', **title_font)
    plt.xticks(rotation=0)
    plt.savefig('GroupAvgShop.png')
    
```

When looking at purchasing habits, nothing really stands out as distinguishing between market segments. However, Married people with PhDs seem to make more purchases overall and take advantage of promotional campaigns, which pairs well with what we just learned about their wine spending habits.
